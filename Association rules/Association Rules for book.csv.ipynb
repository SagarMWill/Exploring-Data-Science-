```python
import pandas as pd
import numpy as np
from mlxtend.frequent_patterns import apriori,association_rules
from mlxtend.preprocessing import TransactionEncoder
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv("book.csv")
```

```python
df.columns
```

```python
df
```

```python
frequent_itemsets_1=apriori(df,min_support=0.1,use_colnames=True)
plt.figure(figsize=(20,10))
sns.barplot('itemsets','support',data=frequent_itemsets_1)
plt.xticks(rotation=90)
```

```python
frequent_itemsets_2=apriori(df,min_support=0.2,use_colnames=True)
sns.barplot('itemsets','support',data=frequent_itemsets_2)
plt.xticks(rotation=90)
```

```python
frequent_itemsets_3=apriori(df,min_support=0.3,use_colnames=True)
frequent_itemsets_3
```

```python
frequent_itemsets_4=apriori(df,min_support=0.4,use_colnames=True)
frequent_itemsets_4
```

```python
frequent_itemsets_5=apriori(df,min_support=0.5,use_colnames=True)
frequent_itemsets_5
```

#### By using diffrenet min_support levels we can see the no of frequent itemsets
* For 10% min_support there is a 39 frequent itemsets.
* For 50% min_support there is no any frequent itemsets.

```python
frequent=association_rules(frequent_itemsets_1,metric='lift',min_threshold=0.75)
frequent
```

```python
frequent[frequent['lift']>2].reset_index(drop=True)
```
