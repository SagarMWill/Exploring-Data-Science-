```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from mlxtend.frequent_patterns import apriori,association_rules
from mlxtend.preprocessing import TransactionEncoder
```

```python
df=pd.read_csv('my_movies.csv')
```

```python
df
```

```python
df_1=df.iloc[:,5:]
```

```python
df_1
```

```python
frequent_itemsets=apriori(df_1,min_support=0.1,use_colnames=True)
plt.figure(figsize=(14,8))
sns.barplot('itemsets','support',data=frequent_itemsets)
plt.xticks(rotation=90)
```

```python
frequent_itemsets_2=apriori(df_1,min_support=0.4,use_colnames=True)
plt.figure(figsize=(14,8))
sns.barplot('itemsets','support',data=frequent_itemsets_2)
plt.xticks(rotation=90)
```

```python
frequent=association_rules(frequent_itemsets_2,metric='lift',min_threshold=1)
frequent
```
